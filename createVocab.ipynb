```python
import pandas as pd
```

* Split dataset into chunks and store word frequencies for each chunk

```python
sample = pd.DataFrame()
round = 0
i = 1
vocabEn = []
vocabFr = []
for df in pd.read_csv("archive/en-fr.csv", chunksize= 10 ** 5):
    df.dropna(inplace=True)
    df['en'] = df['en'].str.lower().str.replace("[^a-z]+", " ", regex=True).str.strip()
    df['fr'] = df['fr'].str.lower().str.replace("[^a-zàâçéèêîôûù]+", " ", regex=True).str.strip()
    vocabEn += df['en'].str.cat(sep=" ").split(" ")
    vocabFr += df['fr'].str.cat(sep=" ").split(" ")
    if i % 20 == 0:
        vocabEn = pd.Series(vocabEn, name='en').value_counts()
        vocabEn.to_csv("vocabEn" + str(round) + ".csv")
        vocabFr = pd.Series(vocabFr, name='fr').value_counts()
        vocabFr.to_csv("vocabFr" + str(round) + ".csv")
        vocabEn = []
        vocabFr = []
        round += 1
    i += 1
vocabEn = pd.Series(vocabEn, name='en').value_counts()
vocabEn.to_csv("vocabEn" + str(round) + ".csv")
vocabFr = pd.Series(vocabFr, name='fr').value_counts()
vocabFr.to_csv("vocabFr" + str(round) + ".csv")
```

* Create the english dictionary by concatenating all word frequency dataframes into one

```python
vocabEn = pd.read_csv("vocabEn0.csv", index_col='en')
for i in range(1, 12):
    vocabEn1 = pd.read_csv("vocabEn" + str(i) + ".csv", index_col='en')
    vocabEn = pd.concat([vocabEn, vocabEn1])
vocabEn = vocabEn.groupby('en').sum().sort_values('count')[::-1]
vocabEn.to_csv("vocabEn.csv")
```

* Do the same for the french dictionary

```python
vocabFr = pd.read_csv("vocabFr0.csv", index_col='fr')
for i in range(1, 12):
    vocabFr1 = pd.read_csv("vocabFr" + str(i) + ".csv", index_col='fr')
    vocabFr = pd.concat([vocabFr, vocabFr1])
vocabFr = vocabFr.groupby('fr').sum().sort_values('count')[::-1]
vocabFr.to_csv("vocabFr.csv")
```
